```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("3. Police Data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#converting stop_date and stop_time to date and time object

data['stop_date'] = pd.to_datetime(data.stop_date)
data['stop_time'] = pd.to_datetime(data.stop_time)
```

```python
data.dtypes
```

```python
# summary based on male and female and their violation

data.pivot_table(data,index=['driver_gender','violation'])
```

```python
pd.pivot_table(data,index=['driver_gender','violation'],values=['driver_age'],aggfunc=[np.max,np.min,np.mean,len])
```

```python
# total number of male and female drivers

data['driver_gender'].value_counts()
```

```python
data['driver_gender'].value_counts().plot.bar()
```

```python
# how many males and females drivers are there for each violation

data.groupby(['violation']).driver_gender.value_counts()
```

```python
# number of people arrested on each date

data.groupby(['stop_date']).is_arrested.value_counts()
```

```python
# on which time police was more active

data['stop_time'].mean()
```

```python
# finding min,max and average age of each male and female

data.groupby('driver_gender').driver_age.agg(['min','max','mean'])
```

```python
data.head()
```

```python
#total true and false cases for drug related activity 

data['drugs_related_stop'].value_counts()
```

```python
# checking drug activity status for both male and female

drugs_output = pd.DataFrame(data.groupby('driver_gender').drugs_related_stop.value_counts())
drugs_output
```

```python
# counts of people for each action taken by police

data['stop_outcome'].value_counts()
```

```python
# counts of people for each action taken by police based on male and female

stop_outcome_result = pd.DataFrame(data.groupby('driver_gender').stop_outcome.value_counts())
stop_outcome_result
```

```python
# what kind of action is taken by police on each date

result = pd.DataFrame(data.groupby('stop_date').stop_outcome.value_counts())
result
```

```python
#number of times police has given warning as stop outcome

warning_data = data[data['stop_outcome']=='Warning']
warning_data
```

```python
warning_data.groupby('driver_gender').stop_outcome.value_counts()
```

```python
arrest_data = data[data['stop_outcome']=='Arrest Driver']
arrest_data
```

```python
# number of people arrested based on different violations

count = arrest_data['violation'].value_counts()
count
```

```python
count.plot(kind='bar')
```

```python
data.groupby('violation').is_arrested.value_counts()
```

```python
sns.countplot(x="is_arrested", hue="violation", data=data);
```

```python
arrest_data['violation'].value_counts()
```

```python
data["driver_gender"].value_counts().plot(kind="pie")
# most people arrested by police were male
```
